## Part 1: Modeling

### 1.1 Problem Description

In our study, we are faced with a vehicle routing problem (Vehicle Routing Problem - VRP).
This problem is a class of operations research and combinatorial optimization problems. It is an extension of the classic traveling salesman problem (TSP), which involves finding the shortest path to connect all vertices of a graph. However, in the case of the VRP, we must determine the routes for a fleet of vehicles that all start and finish at a depot. The objective is to optimize the organization of travel by visiting all vertices of the graph while respecting constraints and distances between locations. In other words, we need to find the best possible routes for our fleet of vehicles to minimize operational costs and ensure maximum efficiency in our logistics operations.


### 1.2 Mathematical Model

#### Indices

- \( I \): Set of cities.
- \( K \): Set of vehicles.

#### Constant Parameters

- \( c_{ij} \): Cost of travel between city i and city j.

#### Decision Variables

- \( x_{ij} \): Indicates 1 if city j is visited from city i, and 0 otherwise.
- \( u_{i} \): Variable used to eliminate sub-tours.

$$
x_{ij} \in \{ 0,1 \}
$$
$$
u_{i} \ge  0
$$

#### Objective Function

$$
\sum_{i \in I} \sum_{j \in I}  c_{ij} \times x_{ij}
$$

#### Constraints:

##### Constraint 1:
Ensures that each node is entered once by a vehicle.

$$
\sum_{i \in I }   x_{ij} = 1 \qquad \forall j \in I \setminus \{1\}
$$

##### Constraint 2:
Ensures the exit of each vehicle and the point from which it enters.

$$
\sum_{i \in I }  x_{ij} = \sum_{i \in I }  x_{ji}  \qquad \forall j \in I
$$

##### Constraint 3:
Each vehicle leaves the depot.

$$
\sum_{j \in I \setminus \{1\}}  x_{1j} = 1
$$

##### Constraint 4:
Eliminate sub-tours using the MillerTuckerZemlin (MTZ) method.

$$
u_{i} - u_{j} + (I - K) \times x_{ij} \leq I - K - 1 \qquad \forall i, j \in I, j \neq 1
$$


### 1.3 Complexity of the Problem*

As a reminder, polynomial reduction is a key concept used to demonstrate that one problem is at least as difficult as another. Specifically, a problem A is polynomially reducible to a problem B if every instance of problem A can be transformed into an instance of problem B in polynomial time. If such a reduction exists, it means that B is at least as difficult as A in terms of computational complexity.

Mathematically, a decision problem A is polynomially reducible to a decision problem B (denoted \( A \leq_p B \)) if there is a function \( f: \Sigma^* \rightarrow \Sigma^* \) that transforms an instance \( x \) of A into an instance \( f(x) \) of B such that:

- \( f(x) \) can be calculated in polynomial time.
- \( x \in A \) if and only if \( f(x) \in B \).

In other words, the function \( f \) translates instances of A into instances of B so that the solution to \( x \) in A is the same as the solution to \( f(x) \) in B.
The Vehicle Routing Problem (VRP) is indeed considered a classic extension of the traveling salesman problem (TSP) and belongs to the class of NP-complete problems.

To prove this, assume we have an instance of the TSP with a set of cities C = {c1, c2, ..., cn} and a cost matrix D, where D[i][j] is the cost to go from city ci to city cj.

Here is a possible reduction function:

- Create a set of cities C' for the VRP that is exactly the same as C.
- Create a cost matrix D' for the VRP that is exactly the same as D.
- Set the number of vehicles for the VRP to 1. This means that we have a single "traveling salesman" in the VRP, as in the TSP.
- Ignore all capacity constraints in the VRP. In the TSP, there are no capacity constraints, so we must ignore them in the VRP for the reduction to be valid.

This reduction function works in polynomial time because copying the set of cities and the cost matrix takes time O(n^2), where n is the number of cities. Therefore, the TSP is reducible to the VRP in polynomial time.

Consequently, if we have an algorithm that can solve the VRP, we can use it to solve the TSP. This shows that the VRP is at least as difficult as the TSP.

Thus, the VRP belongs to the class of NP-complete problems, requiring specific resolution methods to find efficient solutions.

## Demonstration of the NP-Hardness of the TSP Problem

Now, we have performed a polynomial reduction of the TSP problem to our VRP problem, demonstrating that our decision problem is at least as difficult as the TSP problem.
To prove that our problem is NP-hard, we will demonstrate that the decision problem of the traveling salesman is NP-hard.

As a reminder, to show that a problem is NP-hard, we must follow these steps:

1. Choose a known NP-complete witness problem. Our choice is the Hamiltonian cycle problem.
2. Perform a polynomial reduction of the witness problem to our problem.

- **The Hamiltonian Cycle (HC) problem** is a well-known decision problem in complexity theory. It is considered **NP-complete**, meaning it is among the most difficult problems in the **NP** class.

We can observe that:

* If $G$ has a Hamiltonian cycle, then $G'$ has a cycle whose total weight is $|V|$, which is the number of vertices in the graph. Each edge of this cycle has a weight of 1, because each edge is part of the Hamiltonian cycle in $G$.

* If $G'$ has a cycle whose total weight is $|V|$, then each edge of this cycle must have a weight of 1, which means that it was an edge in the original graph $G$. Consequently, this cycle is a Hamiltonian cycle in $G$.

This reduction can be carried out in polynomial time, as it is possible to construct $G'$ and define $w$ in time $O(|V|^2)$, which is polynomial. Therefore, we have shown that **HC $\leq_p$ TSP**, which means that the TSP is at least as difficult as the HC. Given that the HC is NP-hard, this implies that **the TSP decision problem** is also NP-hard.

Voici un exemple basique d'implmentation

Gnrant d'abord les voisins d'une solutions donne, en utilisant une opration simple : Swap, qui consiste  permuter deux sommets de la solution donne:

```python
# Gnration des voisins
def generate_voisins(solution):
    voisins = []
    for i in range(1, len(solution)-1):
        for j in range(i+1, len(solution)):
            voisin = solution.copy()
            voisin[i], voisin[j] = solution[j], solution[i]  # SWAP
            voisins.append(voisin)
    return voisins
```

On peut constater que l'ensemble de voisinage d'une solution est de l'ordre de $n^2$ donc cette fonction s'execute en temps polynomial.

```python
def tabu_search(init_solution, tabu_size, iter_max, distances):
    """
    1. On part d'un lment de notre ensemble de recherche qu'on dclare lment courant
    2. On considre le voisinage de l'lment courant, on choisit le meilleur d'entre
       eux comme nouvel lment courant parmi ceux absents de la liste tabou, et on l'ajoute
        la liste tabou
    3. On boucle jusqu' condition de sortie.
    """
    tabu_list = deque((), maxlen = tabu_size)

    # variables solutions pour la recherche du voisin optimal non tabou
    current_solution = init_solution
    best_solution = init_solution
    best_global_solution = init_solution

    # variables valeurs  pour la recherche du voisin optimal non tabou
    best_cost = calculate_cost(init_solution, distances)
    best_global_cost = best_cost

    # Liste des meilleurs cots  chaque itration
    costs = [best_cost]
    best_global_costs = [best_global_cost]

    for _ in range(iter_max):
        best_cost = float('inf')

        # on parcourt tous les voisins de la solution courante
        neighbors = generate_voisins(current_solution)
        for neighbor in neighbors:
            neighbor_cost = calculate_cost(neighbor, distances)
            # meilleure solution non taboue trouve
            if neighbor_cost < best_cost and neighbor not in tabu_list:
                best_cost = neighbor_cost
                best_solution = neighbor

        # on met  jour la meilleure solution rencontre depuis le dbut
        if best_cost < best_global_cost:
            best_global_cost = best_cost
            best_global_solution = best_solution

        costs.append(best_cost)
        best_global_costs.append(best_global_cost)

        # on passe au meilleur voisin non tabou trouv
        current_solution = best_solution

        # on met  jour la liste tabou
        tabu_list.append(current_solution)

    return best_global_solution, costs, best_global_costs



def multi_start_tabu_search(init_solutions, tabu_size, iter_max, distances):
    all_costs = []  # Initialiser la liste pour contenir tous les cots
    best_solutions = []
    for init_solution in init_solutions:
        best_solution, costs, best_global_cost = tabu_search(init_solution, tabu_size, iter_max, distances)
        best_solutions.append(best_solution)
        all_costs.append(costs)  # Ajouter les cots de cette solution  all_costs

    # Affichage pour le dbogage
    print(f"best_solutions : {best_solutions}")

    # Retourner la meilleure solution parmi toutes les solutions trouves, et tous les cots
    return min(best_solutions, key=lambda x: calculate_cost(x, distances)), all_costs



# Gnrer plusieurs solutions initiales
init_solutions = [generate_solution(matrice_distance, nbvehicule) for _ in range(nb_multistart)]
# Appliquer la recherche tabou multi-start
best_solution = multi_start_tabu_search(init_solutions, tailleTabu, iterMaximum, matrice_distance)
# Afficher la meilleure solution
print("La meilleure solution : \n" ,best_solution)
# visualiser_solution(best_solution, coords)
```

Commanant par le calcul de la borne infrieure pour un scnario donne :

```python
import numpy as np
from pulp import *


def borne_inferieur(graphe, nb_villes, nb_vehicules):
    # variables
    x = [[[pulp.LpVariable(f"x{i}_{j},{k}", cat="Binary") if i != j else None for k in range(nb_vehicules)] for j in range(nb_villes)] for i in range(nb_villes)]
    # probleme
    prob = LpProblem("vrp", LpMinimize)

    # fonction objective
    cost = lpSum(graphe[i][j] * x[i][j][k] if i != j else 0 for k in range(nb_vehicules) for j in range(nb_villes) for i in range (nb_villes))
    prob += cost

    # contraintes
    for j in range(nb_villes):
        prob +=lpSum(x[i][j][k] if i != j else 0 for i in range(0,nb_villes) for k in range(nb_vehicules)) == 1

    for k in range(nb_vehicules):
        prob += lpSum(x[0][j][k] for j in range(nb_villes)) == 1
        prob += lpSum(x[i][0][k] for i in range(nb_villes)) == 1

    for k in range(nb_vehicules):
        for j in range(nb_villes):
            prob += lpSum(x[i][j][k] if i != j else 0 for i in range(nb_villes)) -  lpSum(x[j][i][k] for i in range(nb_villes)) == 0

    prob.solve()
    return value(prob.objective)

# Utilisez la fonction avec vos propres paramtres
graphe = matrice_distance
nb_villes = 10
nb_vehicules = 3

borne = borne_inferieur(graphe, nb_villes, nb_vehicules)
print(borne)
```

## Implmentation :
Prenant un exemple d'une instance donne avec 1 seule camion:



```python
import random
import matplotlib.pyplot as plt

# Donnes du problme
CITY_COORDINATES = [[5, 80], [124, 31], [46, 54], [86, 148], [21, 8],
                   [134, 72], [49, 126], [36, 34], [26, 49], [141, 6],
                   [124, 122], [80, 92], [70, 69], [76, 133], [23, 65]]

DEPOT_INDEX = 0  # Indice de la ville de dpart (dpt)
POPULATION_SIZE = 100
MAX_GENERATION = 200
CROSSOVER_RATE = 0.9
MUTATION_RATE = 0.1
ELITE_SIZE = 5  # Taille de l'lite

# Calculer la matrice des distances entre les villes
DISTANCE_MATRIX = [[0.0] * len(CITY_COORDINATES) for _ in range(len(CITY_COORDINATES))]
for i in range(len(CITY_COORDINATES)):
    for j in range(len(CITY_COORDINATES)):
        x1, y1 = CITY_COORDINATES[i]
        x2, y2 = CITY_COORDINATES[j]
        distance = ((x1 - x2) ** 2 + (y1 - y2) ** 2) ** 0.5
        DISTANCE_MATRIX[i][j] = distance


def create_initial_population(size):
    population = []
    for _ in range(size):
        # Crer un individu alatoire (chemin)
        individual = list(range(1, len(CITY_COORDINATES)))  # Les indices des villes (hors dpt)
        random.shuffle(individual)
        population.append(individual)
    return population


def evaluate_fitness(individual):
    # Calcule la distance totale du parcours de l'individu (chemin)
    distance = DISTANCE_MATRIX[DEPOT_INDEX][individual[0]]
    for i in range(len(individual) - 1):
        distance += DISTANCE_MATRIX[individual[i]][individual[i + 1]]
    distance += DISTANCE_MATRIX[individual[-1]][DEPOT_INDEX]
    return distance


def tournament_selection(population, k):
    # Slectionne k individus alatoires de la population et retourne le meilleur
    participants = random.sample(population, k)
    return min(participants, key=evaluate_fitness)


def order_crossover(parent1, parent2):
    # Cre un enfant par crossover (croisement d'ordre) de deux parents
    start_index = random.randint(0, len(parent1) - 2)
    end_index = random.randint(start_index + 1, len(parent1) - 1)
    child = [-1] * len(parent1)
    child[start_index:end_index + 1] = parent1[start_index:end_index + 1]
    remaining_values = [value for value in parent2 if value not in child]
    j = 0
    for i in range(len(child)):
        if child[i] == -1:
            child[i] = remaining_values[j]
            j += 1
    return child


def mutate(individual):
    # Effectue une mutation alatoire sur l'individu (change de deux positions)
    index1 = random.randint(0, len(individual) - 1)
    index2 = random.randint(0, len(individual) - 1)
    individual[index1], individual[index2] = individual[index2], individual[index1]


def evolve_population(population):
    new_population = []

    # Elitisme : Conserver les meilleurs individus
    elite = sorted(population, key=evaluate_fitness)[:ELITE_SIZE]
    new_population.extend(elite)

    while len(new_population) < len(population):
        parent1 = tournament_selection(population, k=5)
        if random.random() < CROSSOVER_RATE:
            parent2 = tournament_selection(population, k=5)
            child = order_crossover(parent1, parent2)
        else:
            child = parent1
        if random.random() < MUTATION_RATE:
            mutate(child)
        new_population.append(child)

    return new_population


def find_best_solution(population):
    best_individual = min(population, key=evaluate_fitness)
    best_fitness = evaluate_fitness(best_individual)
    return best_individual, best_fitness


def visualize_solution(solution):
    # Affiche le parcours de la meilleure solution trouve
    x = [CITY_COORDINATES[DEPOT_INDEX][0]]
    y = [CITY_COORDINATES[DEPOT_INDEX][1]]
    for city_index in solution:
        x.append(CITY_COORDINATES[city_index][0])
        y.append(CITY_COORDINATES[city_index][1])
    x.append(CITY_COORDINATES[DEPOT_INDEX][0])
    y.append(CITY_COORDINATES[DEPOT_INDEX][1])
    plt.plot(x, y, marker='o')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('VRP Solution')
    plt.show()


def main():
    population = create_initial_population(POPULATION_SIZE)
    best_solution, best_fitness = find_best_solution(population)
    all_fitness = [best_fitness]

    for generation in range(1, MAX_GENERATION + 1):
        population = evolve_population(population)
        best_solution, best_fitness = find_best_solution(population)
        all_fitness.append(best_fitness)
        print(f"Generation: {generation}, Best Fitness: {best_fitness}")

    print(f"\nBest Solution: {best_solution}")
    print(f"Best Fitness: {best_fitness}")

    visualize_solution(best_solution)

    plt.plot(all_fitness)
    plt.xlabel('Generation')
    plt.ylabel('Best Fitness')
    plt.title('Fitness Progress')
    plt.show()


if __name__ == '__main__':
    main()

```
