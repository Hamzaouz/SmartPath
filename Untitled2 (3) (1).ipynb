## Part 1: Modeling

### 1.1 Problem Description

In our study, we are faced with a vehicle routing problem (Vehicle Routing Problem - VRP).
This problem is a class of operations research and combinatorial optimization problems. It is an extension of the classic traveling salesman problem (TSP), which involves finding the shortest path to connect all vertices of a graph. However, in the case of the VRP, we must determine the routes for a fleet of vehicles that all start and finish at a depot. The objective is to optimize the organization of travel by visiting all vertices of the graph while respecting constraints and distances between locations. In other words, we need to find the best possible routes for our fleet of vehicles to minimize operational costs and ensure maximum efficiency in our logistics operations.


### 1.2 Mathematical Model

#### Indices

- \( I \): Set of cities.
- \( K \): Set of vehicles.

#### Constant Parameters

- \( c_{ij} \): Cost of travel between city i and city j.

#### Decision Variables

- \( x_{ij} \): Indicates 1 if city j is visited from city i, and 0 otherwise.
- \( u_{i} \): Variable used to eliminate sub-tours.

$$
x_{ij} \in \{ 0,1 \}
$$
$$
u_{i} \ge  0
$$

#### Objective Function

$$
\sum_{i \in I} \sum_{j \in I}  c_{ij} \times x_{ij}
$$

#### Constraints:

##### Constraint 1:
Ensures that each node is entered once by a vehicle.

$$
\sum_{i \in I }   x_{ij} = 1 \qquad \forall j \in I \setminus \{1\}
$$

##### Constraint 2:
Ensures the exit of each vehicle and the point from which it enters.

$$
\sum_{i \in I }  x_{ij} = \sum_{i \in I }  x_{ji}  \qquad \forall j \in I
$$

##### Constraint 3:
Each vehicle leaves the depot.

$$
\sum_{j \in I \setminus \{1\}}  x_{1j} = 1
$$

##### Constraint 4:
Eliminate sub-tours using the Miller–Tucker–Zemlin (MTZ) method.

$$
u_{i} - u_{j} + (I - K) \times x_{ij} \leq I - K - 1 \qquad \forall i, j \in I, j \neq 1
$$


### 1.3 Complexity of the Problem*

As a reminder, polynomial reduction is a key concept used to demonstrate that one problem is at least as difficult as another. Specifically, a problem A is polynomially reducible to a problem B if every instance of problem A can be transformed into an instance of problem B in polynomial time. If such a reduction exists, it means that B is at least as difficult as A in terms of computational complexity.

Mathematically, a decision problem A is polynomially reducible to a decision problem B (denoted \( A \leq_p B \)) if there is a function \( f: \Sigma^* \rightarrow \Sigma^* \) that transforms an instance \( x \) of A into an instance \( f(x) \) of B such that:

- \( f(x) \) can be calculated in polynomial time.
- \( x \in A \) if and only if \( f(x) \in B \).

In other words, the function \( f \) translates instances of A into instances of B so that the solution to \( x \) in A is the same as the solution to \( f(x) \) in B.
The Vehicle Routing Problem (VRP) is indeed considered a classic extension of the traveling salesman problem (TSP) and belongs to the class of NP-complete problems.

To prove this, assume we have an instance of the TSP with a set of cities C = {c1, c2, ..., cn} and a cost matrix D, where D[i][j] is the cost to go from city ci to city cj.

Here is a possible reduction function:

- Create a set of cities C' for the VRP that is exactly the same as C.
- Create a cost matrix D' for the VRP that is exactly the same as D.
- Set the number of vehicles for the VRP to 1. This means that we have a single "traveling salesman" in the VRP, as in the TSP.
- Ignore all capacity constraints in the VRP. In the TSP, there are no capacity constraints, so we must ignore them in the VRP for the reduction to be valid.

This reduction function works in polynomial time because copying the set of cities and the cost matrix takes time O(n^2), where n is the number of cities. Therefore, the TSP is reducible to the VRP in polynomial time.

Consequently, if we have an algorithm that can solve the VRP, we can use it to solve the TSP. This shows that the VRP is at least as difficult as the TSP.

Thus, the VRP belongs to the class of NP-complete problems, requiring specific resolution methods to find efficient solutions.

## Demonstration of the NP-Hardness of the TSP Problem

Now, we have performed a polynomial reduction of the TSP problem to our VRP problem, demonstrating that our decision problem is at least as difficult as the TSP problem.
To prove that our problem is NP-hard, we will demonstrate that the decision problem of the traveling salesman is NP-hard.

As a reminder, to show that a problem is NP-hard, we must follow these steps:

1. Choose a known NP-complete witness problem. Our choice is the Hamiltonian cycle problem.
2. Perform a polynomial reduction of the witness problem to our problem.

- **The Hamiltonian Cycle (HC) problem** is a well-known decision problem in complexity theory. It is considered **NP-complete**, meaning it is among the most difficult problems in the **NP** class.

We can observe that:

* If $G$ has a Hamiltonian cycle, then $G'$ has a cycle whose total weight is $|V|$, which is the number of vertices in the graph. Each edge of this cycle has a weight of 1, because each edge is part of the Hamiltonian cycle in $G$.

* If $G'$ has a cycle whose total weight is $|V|$, then each edge of this cycle must have a weight of 1, which means that it was an edge in the original graph $G$. Consequently, this cycle is a Hamiltonian cycle in $G$.

This reduction can be carried out in polynomial time, as it is possible to construct $G'$ and define $w$ in time $O(|V|^2)$, which is polynomial. Therefore, we have shown that **HC $\leq_p$ TSP**, which means that the TSP is at least as difficult as the HC. Given that the HC is NP-hard, this implies that **the TSP decision problem** is also NP-hard.


Our optimization problem is more difficult than the corresponding decision problem, since an algorithm that finds the smallest cost value required to visit each city can be used to answer the decision problem (if $k > \min$, the answer is "No", otherwise it's "Yes"). # Métaheuristiques :



**Une métaheuristique** est un algorithme d’optimisation visant à résoudre des problèmes d’optimisation difficile (souvent issus des domaines de la recherche opérationnelle, de l'ingénierie ou de l'intelligence artificielle) pour lesquels on ne connaît pas de méthode classique plus efficace.

Les métaheuristiques sont généralement des algorithmes stochastiques itératifs, qui progressent vers un optimum global, c'est-à-dire l'extremum global d'une fonction, par échantillonnage d’une fonction objectif. Elles se comportent comme des algorithmes de recherche, tentant d’apprendre les caractéristiques d’un problème afin d’en trouver une approximation de la meilleure solution (d'une manière proche des algorithmes d'approximation).
On parle de méta, du grec « au-delà » (comprendre ici « à un plus haut niveau »), heuristique, du grec ευρισκειν / heuriskein, qui signifie « trouver ». En effet, ces algorithmes se veulent des méthodes génériques pouvant optimiser une large gamme de problèmes différents, sans nécessiter de changements profonds dans l’algorithme employé.



## Types des méta-heuristiques
Il ya plusieurs types de métaheuristiques. Nous dans notre travail, on a utlisé l’algorithme Génétique et l’algorithme Tabou pour résoudre le VRP.

###Recherche Tabou
L’algorithme tabou est une métaheuristique d’optimisation présentée par Fred W. Glover en 1986. Cette méthode est une métaheuristique itérative qualifiée de recherche locale au sens large. Le principe de la recherche Tabou repose sur une méthode de déplacement sur l’espace des solutions, tout en cherchant constamment à améliorer la meilleure solution courante et en conservant en mémoire la liste des précédents déplacements et ainsi guider la recherche en dehors de zones précédemment parcourues.



image.png

###Algorithme génétique
Un algorithme génétique est une méthode de résolution de problèmes d’optimisation, avec ou sans contraintes, basée sur un processus de sélection naturelle (processus analogue à celui de l’évolution biologique). Dans un tel algorithme, une population de solutions est modifiée à plusieurs reprises. À chaque fois, l’algorithme sélectionne les solutions les plus adaptées et les croise entre elles pour créer de nouvelles solutions.

## Choix de la méthode de recherche tabou

Après avoir exploré plusieurs méta-heuristiques, nous avons décidé d'adopter l'algorithme Tabu pour aborder notre problème d'optimisation. L'algorithme Tabu est particulièrement efficace pour des problèmes qui présentent un grand nombre de solutions potentielles, où le risque de se retrouver coincé dans des optima locaux est élevé. Par rapport à d'autres méta-heuristiques, l'algorithme Tabu offre un mécanisme robuste d'exploration et d'exploitation de l'espace des solutions. Il est capable d'échapper à la captivité d'un optimum local en utilisant sa liste tabou, qui interdit temporairement le retour à des solutions précédemment visitées.

De plus, contrairement à certaines autres méta-heuristiques, l'algorithme Tabu ne nécessite pas de gradient ou d'autres informations dérivées, ce qui le rend applicable à une plus grande variété de problèmes, y compris ceux où ces informations peuvent être difficiles à obtenir ou à calculer. Par ailleurs, l'algorithme Tabu est hautement paramétrable, ce qui permet d'affiner sa performance en fonction des spécificités du problème à résoudre.

Finalement, l'algorithme Tabu a été largement utilisé et validé dans diverses applications réelles, démontrant ainsi sa capacité à résoudre efficacement une variété de problèmes d'optimisation. C'est pourquoi, en tenant compte de la complexité et des caractéristiques de notre problème, nous avons choisi de mettre en œuvre l'algorithme Tabu comme principale stratégie de recherche et d'optimisation.

Voici un exemple basique d'implémentation

Générant d'abord les voisins d'une solutions donnée, en utilisant une opération simple : Swap, qui consiste à permuter deux sommets de la solution donnée:

```python
# Génération des voisins
def generate_voisins(solution):
    voisins = []
    for i in range(1, len(solution)-1):
        for j in range(i+1, len(solution)):
            voisin = solution.copy()
            voisin[i], voisin[j] = solution[j], solution[i]  # SWAP
            voisins.append(voisin)
    return voisins
```

On peut constater que l'ensemble de voisinage d'une solution est de l'ordre de $n^2$ donc cette fonction s'execute en temps polynomial.

Voici notre algorithme tabou simple et avec multi-start :

```python
def tabu_search(init_solution, tabu_size, iter_max, distances):
    """
    1. On part d'un élément de notre ensemble de recherche qu'on déclare élément courant
    2. On considère le voisinage de l'élément courant, on choisit le meilleur d'entre
       eux comme nouvel élément courant parmi ceux absents de la liste tabou, et on l'ajoute
       à la liste tabou
    3. On boucle jusqu'à condition de sortie.
    """
    tabu_list = deque((), maxlen = tabu_size)

    # variables solutions pour la recherche du voisin optimal non tabou
    current_solution = init_solution
    best_solution = init_solution
    best_global_solution = init_solution

    # variables valeurs  pour la recherche du voisin optimal non tabou
    best_cost = calculate_cost(init_solution, distances)
    best_global_cost = best_cost

    # Liste des meilleurs coûts à chaque itération
    costs = [best_cost]
    best_global_costs = [best_global_cost]

    for _ in range(iter_max):
        best_cost = float('inf')

        # on parcourt tous les voisins de la solution courante
        neighbors = generate_voisins(current_solution)
        for neighbor in neighbors:
            neighbor_cost = calculate_cost(neighbor, distances)
            # meilleure solution non taboue trouvée
            if neighbor_cost < best_cost and neighbor not in tabu_list:
                best_cost = neighbor_cost
                best_solution = neighbor

        # on met à jour la meilleure solution rencontrée depuis le début
        if best_cost < best_global_cost:
            best_global_cost = best_cost
            best_global_solution = best_solution

        costs.append(best_cost)
        best_global_costs.append(best_global_cost)

        # on passe au meilleur voisin non tabou trouvé
        current_solution = best_solution

        # on met à jour la liste tabou
        tabu_list.append(current_solution)

    return best_global_solution, costs, best_global_costs



def multi_start_tabu_search(init_solutions, tabu_size, iter_max, distances):
    all_costs = []  # Initialiser la liste pour contenir tous les coûts
    best_solutions = []
    for init_solution in init_solutions:
        best_solution, costs, best_global_cost = tabu_search(init_solution, tabu_size, iter_max, distances)
        best_solutions.append(best_solution)
        all_costs.append(costs)  # Ajouter les coûts de cette solution à all_costs

    # Affichage pour le débogage
    print(f"best_solutions : {best_solutions}")

    # Retourner la meilleure solution parmi toutes les solutions trouvées, et tous les coûts
    return min(best_solutions, key=lambda x: calculate_cost(x, distances)), all_costs



# Générer plusieurs solutions initiales
init_solutions = [generate_solution(matrice_distance, nbvehicule) for _ in range(nb_multistart)]
# Appliquer la recherche tabou multi-start
best_solution = multi_start_tabu_search(init_solutions, tailleTabu, iterMaximum, matrice_distance)
# Afficher la meilleure solution
print("La meilleure solution : \n" ,best_solution)
# visualiser_solution(best_solution, coords)
```

Cependant, il est important de noter que la méthode du simplexe s'applique aux problèmes de programmation linéaire continus. Dans notre cas, le problème du VRP (Vehicle Routing Problem), nous travaillons avec des variables discrètes (les villes ne peuvent pas être visitées de manière fractionnaire), ce qui rend l'application directe du simplexe impossible. Malgré cela, la borne inférieure obtenue par cette méthode reste une référence précieuse pour évaluer la performance de notre algorithme.

La distance entre les solutions générées par notre algorithme et la borne inférieure sert de mesure de la qualité des solutions. Plus notre solution est proche de la borne inférieure, meilleure est sa qualité.

Commançant par le calcul de la borne inférieure pour un scénario donnée :

```python
import numpy as np
from pulp import *


def borne_inferieur(graphe, nb_villes, nb_vehicules):
    # variables
    x = [[[pulp.LpVariable(f"x{i}_{j},{k}", cat="Binary") if i != j else None for k in range(nb_vehicules)] for j in range(nb_villes)] for i in range(nb_villes)]
    # probleme
    prob = LpProblem("vrp", LpMinimize)

    # fonction objective
    cost = lpSum(graphe[i][j] * x[i][j][k] if i != j else 0 for k in range(nb_vehicules) for j in range(nb_villes) for i in range (nb_villes))
    prob += cost

    # contraintes
    for j in range(nb_villes):
        prob +=lpSum(x[i][j][k] if i != j else 0 for i in range(0,nb_villes) for k in range(nb_vehicules)) == 1

    for k in range(nb_vehicules):
        prob += lpSum(x[0][j][k] for j in range(nb_villes)) == 1
        prob += lpSum(x[i][0][k] for i in range(nb_villes)) == 1

    for k in range(nb_vehicules):
        for j in range(nb_villes):
            prob += lpSum(x[i][j][k] if i != j else 0 for i in range(nb_villes)) -  lpSum(x[j][i][k] for i in range(nb_villes)) == 0

    prob.solve()
    return value(prob.objective)

# Utilisez la fonction avec vos propres paramètres
graphe = matrice_distance
nb_villes = 10
nb_vehicules = 3

borne = borne_inferieur(graphe, nb_villes, nb_vehicules)
print(borne)
```

## Piste d'évolution
Il y a d'autre méta-heuristique interressant utilisable sur notre problème, un algorithme populaire est l'algorithme génetique (par population)

#### Génétique

Un algorithme génétique est un algorithme appartenant à la famille des algorithmes évolutionnistes. Le but de ces algorithmes est de trouver des solutions approchées pour des problèmes d'optimisations. Ils permettent de fournir une réponse en un temps acceptable lorsque l'on ne connaît pas de méthodes de résolution plus efficaces.

On retrouve la notion de **sélection naturelle** dans ces algorithmes. C'est-à-dire qu'à la fin d'une *génération*, on va garder seulement les *individus* possédant les *valeurs d'aptitudes* les plus hautes. On aura alors une nouvelle génération dans laquelle seuls les individus les plus intéressants auront survécu et fusionnés. On pourra alors, au fur et à mesure des générations, avoir une précision de plus en plus précise.

<img style="margin: 40px auto;" src="https://khayyam.developpez.com/articles/algo/genetic/images/schema_gen.gif" alt="schéma algo génétique" width="300"/>

cet algorithme utilise la théorie de Darwin sur l'évolution des éspèces. On retrouve trois principes :
- **Variation** : Chaque individu est unique. Les différences vont êtres décisives pour le processus de sélection.
- **Adaptation** : Les individus qui arrivent à s'adapter à l'environnement survivront plus facilement. Ceux qui survivent le plus facilement auront des facilités pour se reproduire.
- **Hérédité** : Les caractéristiques d'un individu sont transmises à leur descendance. Ce mécanisme permettra de faire évoluer l'espèce pour partager les caractéristiques facilitant la survie.

Nous allons caractériser une population selon une notion de cercle concentrique comme suit :

<img style="margin: 40px auto;" src="http://igm.univ-mlv.fr/~dr/XPOSE2013/tleroux_genetic_algorithm/slide/xpose_files/PPMf3MP2.png" alt="schéma algo génétique" width="300"/>


- Une **population** correspond à l'ensemble des solutions envisageables
- Un **individu** représente une solution
- Un **chromosome** est une composante de la solution
- Un **gène** se réfère à une caractéristique, une particularité

Grâce à cela, nous pouvons définir trois opérateurs d'évolution :
- La **sélection** : Choix des individus les plus adaptés au problème.
    - Cette transformation permet de converger vers un optimum global. C'est le principe d'*adaptation* de la théorie de Darwin. On a plusieurs sélection possibles :
        - Par rang : Sélection des individus ayant les meilleurs score d'adaptation.
        - Par probabilité de sélection proportionnelle à l'adaptation : La probabilité d'être sélectionner est proportionnelle à son adaptation au problème.
        - Par tournois : Sélection proportionnelle sur des paires d'individus, puis on choisi ceux qui ont le meilleur score d'adaptation.
        - Uniforme : Sélection aléatoire, uniforme et sans valeur d'adaptation.
    - On réalise un croisement entre deux chromosomes parmi la population restante.
- Le **croisement** : Mélange par la reproduction des particularité d'un individu.
    - C'est le résultat obtenu lorsque deux chromosomes partagent leurs particularités. C'est le principe d'*hérédité* de la théorie de Darwin. Il existe deux méthodes de croisement :
        - *Simple* enjambement : Fusionner les particularités de deux individus à partir d'un pivot dans le but d'obtenir un ou deux enfants.

 <img style="margin: 40px auto;" src="https://i0.wp.com/ledatascientist.com/wp-content/uploads/2020/10/One-point-XO.png?resize=768%2C209&ssl=1" alt="schéma algo génétique" width="300">


- *Double* enjambement : Même principe mais avec deux pivots.

<img style="margin: 40px auto;" src="https://i0.wp.com/ledatascientist.com/wp-content/uploads/2020/10/Two-point-XO.png?resize=768%2C209&ssl=1" alt="schéma algo génétique" width="300"/>


 On réalise ensuite une mutation sur les enfants obtenues lors du croisement.
- La **mutation** : Altération aléatoire des particularité d'un individu.
    - On altère un gène dans un chromosome selon un facteur de mutation. C'est la probabilité qu'une mutation soit effectuée sur un individu. C'est le principe de *variation* de la théorie de Darwin. Cela permet aussi de converger vers un extermum local.
        - Insertion : On insert un chromosome à la suite d'un autre.


 <img style="margin: 40px auto;" src="https://i0.wp.com/ledatascientist.com/wp-content/uploads/2020/10/Insertion-mutation.png?resize=768%2C82&ssl=1" alt="schéma algo génétique" width="300"/>
        - Swap : On échange la position de deux chromosomes.
        <img style="margin: 40px auto;" src="https://i0.wp.com/ledatascientist.com/wp-content/uploads/2020/10/Swap-mutation.png?resize=768%2C82&ssl=1" alt="schéma algo génétique" width="300"/>
        - Reversion : On inverse l'ordre des éléments sélectionnés.
        <img style="margin: 40px auto;" src="https://i0.wp.com/ledatascientist.com/wp-content/uploads/2020/10/Reversion-mutation.png?resize=768%2C82&ssl=1" alt="schéma algo génétique" width="300"/>

Enfin, on va créer la prochaine génération. On applique différentes transformations génétiques sur une partie de notre génération actuelle. On dispose alors de deux méthodes :
- **Élitisme** : On sélectionne seulement l'élite et on les positionne dans l'ordre de notre prochain génération. Le reste de la population sera croisé et muté puis ajouté dans l'ordre.

    <img style="margin: 40px auto;" src="https://i0.wp.com/ledatascientist.com/wp-content/uploads/2020/10/Elitisme.png?resize=768%2C326&ssl=1" alt="schéma algo génétique" width="300"/>

- **Fusion** : On prend toute la génération actuelle et on lui ajoute des individus croisés et mutés issus de la génération actuelle. On trie et selectionne les $P$ premiers individus de notre liste pour former la prochaine génération.

    <img style="margin-top: 40px;" src="https://i0.wp.com/ledatascientist.com/wp-content/uploads/2020/10/Fusion.png?resize=768%2C509&ssl=1" alt="schéma algo génétique" width="300"/>

## Etapes pour implementer notre programme génetique

Considerant qu'on a un seul camion, alors la création d'une fonction génétique pour résoudre ce problème plusieurs étapes. Voici les étapes générales:

1/ Représentation des individus : la représentation la plus courante est l'utilisation de permutations, où chaque individu est une séquence ordonnée de villes à visiter. Cette représentation permet de garantir que chaque ville est visitée exactement une fois.

2/ Génération initiale de la population : Une population initiale variée est importante car elle permet d'explorer différentes régions de l'espace de recherche.

3/ Évaluation de la fonction de fitness : L'objectif est de minimiser la longueur totale du parcours, ce qui correspond à trouver le parcours le plus court.

4/ Sélection des parents : le but est de favoriser les individus les plus performants afin de transmettre leurs caractéristiques génétiques à la génération suivante.

5/ Croisement (crossover) : Le but est de combiner les caractéristiques des deux parents afin de créer une descendance prometteuse.

6/ Mutation :  La probabilité de mutation doit être réglée avec soin, car une trop faible valeur peut entraîner une convergence prématurée vers un optimum local, tandis qu'une trop forte valeur peut causer une exploration trop aléatoire.

7/ Remplacement de la population :  Différentes stratégies de remplacement peuvent être utilisées, telles que le remplacement des pires individus ou le remplacement des individus aléatoires.

Ces étapes sont générales et il existe de nombreuses variantes et améliorations possibles pour résoudre efficacement le problème à l'aide d'un algorithme génétique.

## Implémentation :
Prenant un exemple d'une instance donnée avec 1 seule camion:



```python
import random
import matplotlib.pyplot as plt

# Données du problème
CITY_COORDINATES = [[5, 80], [124, 31], [46, 54], [86, 148], [21, 8],
                   [134, 72], [49, 126], [36, 34], [26, 49], [141, 6],
                   [124, 122], [80, 92], [70, 69], [76, 133], [23, 65]]

DEPOT_INDEX = 0  # Indice de la ville de départ (dépôt)
POPULATION_SIZE = 100
MAX_GENERATION = 200
CROSSOVER_RATE = 0.9
MUTATION_RATE = 0.1
ELITE_SIZE = 5  # Taille de l'élite

# Calculer la matrice des distances entre les villes
DISTANCE_MATRIX = [[0.0] * len(CITY_COORDINATES) for _ in range(len(CITY_COORDINATES))]
for i in range(len(CITY_COORDINATES)):
    for j in range(len(CITY_COORDINATES)):
        x1, y1 = CITY_COORDINATES[i]
        x2, y2 = CITY_COORDINATES[j]
        distance = ((x1 - x2) ** 2 + (y1 - y2) ** 2) ** 0.5
        DISTANCE_MATRIX[i][j] = distance


def create_initial_population(size):
    population = []
    for _ in range(size):
        # Créer un individu aléatoire (chemin)
        individual = list(range(1, len(CITY_COORDINATES)))  # Les indices des villes (hors dépôt)
        random.shuffle(individual)
        population.append(individual)
    return population


def evaluate_fitness(individual):
    # Calcule la distance totale du parcours de l'individu (chemin)
    distance = DISTANCE_MATRIX[DEPOT_INDEX][individual[0]]
    for i in range(len(individual) - 1):
        distance += DISTANCE_MATRIX[individual[i]][individual[i + 1]]
    distance += DISTANCE_MATRIX[individual[-1]][DEPOT_INDEX]
    return distance


def tournament_selection(population, k):
    # Sélectionne k individus aléatoires de la population et retourne le meilleur
    participants = random.sample(population, k)
    return min(participants, key=evaluate_fitness)


def order_crossover(parent1, parent2):
    # Crée un enfant par crossover (croisement d'ordre) de deux parents
    start_index = random.randint(0, len(parent1) - 2)
    end_index = random.randint(start_index + 1, len(parent1) - 1)
    child = [-1] * len(parent1)
    child[start_index:end_index + 1] = parent1[start_index:end_index + 1]
    remaining_values = [value for value in parent2 if value not in child]
    j = 0
    for i in range(len(child)):
        if child[i] == -1:
            child[i] = remaining_values[j]
            j += 1
    return child


def mutate(individual):
    # Effectue une mutation aléatoire sur l'individu (échange de deux positions)
    index1 = random.randint(0, len(individual) - 1)
    index2 = random.randint(0, len(individual) - 1)
    individual[index1], individual[index2] = individual[index2], individual[index1]


def evolve_population(population):
    new_population = []

    # Elitisme : Conserver les meilleurs individus
    elite = sorted(population, key=evaluate_fitness)[:ELITE_SIZE]
    new_population.extend(elite)

    while len(new_population) < len(population):
        parent1 = tournament_selection(population, k=5)
        if random.random() < CROSSOVER_RATE:
            parent2 = tournament_selection(population, k=5)
            child = order_crossover(parent1, parent2)
        else:
            child = parent1
        if random.random() < MUTATION_RATE:
            mutate(child)
        new_population.append(child)

    return new_population


def find_best_solution(population):
    best_individual = min(population, key=evaluate_fitness)
    best_fitness = evaluate_fitness(best_individual)
    return best_individual, best_fitness


def visualize_solution(solution):
    # Affiche le parcours de la meilleure solution trouvée
    x = [CITY_COORDINATES[DEPOT_INDEX][0]]
    y = [CITY_COORDINATES[DEPOT_INDEX][1]]
    for city_index in solution:
        x.append(CITY_COORDINATES[city_index][0])
        y.append(CITY_COORDINATES[city_index][1])
    x.append(CITY_COORDINATES[DEPOT_INDEX][0])
    y.append(CITY_COORDINATES[DEPOT_INDEX][1])
    plt.plot(x, y, marker='o')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('VRP Solution')
    plt.show()


def main():
    population = create_initial_population(POPULATION_SIZE)
    best_solution, best_fitness = find_best_solution(population)
    all_fitness = [best_fitness]

    for generation in range(1, MAX_GENERATION + 1):
        population = evolve_population(population)
        best_solution, best_fitness = find_best_solution(population)
        all_fitness.append(best_fitness)
        print(f"Generation: {generation}, Best Fitness: {best_fitness}")

    print(f"\nBest Solution: {best_solution}")
    print(f"Best Fitness: {best_fitness}")

    visualize_solution(best_solution)

    plt.plot(all_fitness)
    plt.xlabel('Generation')
    plt.ylabel('Best Fitness')
    plt.title('Fitness Progress')
    plt.show()


if __name__ == '__main__':
    main()

```

## Resultats
Dans notre cas, nous pouvous voir que la valeur de "Best Fitness" diminue au fil des générations, ce qui suggère que l'algorithme ou la méthode d'optimisation progresse dans la recherche de la meilleure solution.

Cependant, à partir de la génération 7, la meilleure fitness semble se stabiliser autour de 630, avec quelques légères fluctuations. Puis, à partir de la génération 30, il y a une nouvelle diminution jusqu'à atteindre 620 à la génération 30.
